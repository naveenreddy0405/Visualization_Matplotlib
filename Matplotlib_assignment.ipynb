```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#1.Load your company_sales_data.csv CSV file
path = 'company_sales_data.csv'
csv_data = pd.read_csv(path)
csv_data
#csv_data.isnull()
```

```python
#2.Read Total profit of all months and show it using a line plot

#Getting data and converting to List
months = csv_data['month_number'].tolist()
profit = csv_data['total_profit'].tolist()
#print(months,'<<>>' ,profit )

plt.plot(months, profit, label = 'Total profit of all months')
plt.xlabel('Month number')
plt.ylabel('Total profit of all months')
plt.xticks(months)
plt.title('Total profit of all months')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
print('Analysis : The plot shows Total Profit of all Months.Lowest profit was recorded in the 2nd month and highest in 10th month,There is overall growth in profits from month 1 to 12' ) 
```

```python
#3.Generated line plot must include following Style properties: â€“

#Line Style dotted and Line-color should be red
#Add a circle marker.
#Line marker color as read
#Line width should be 3
plt.plot(months, profit, label = 'Profit data of last year',color='r', marker='o', markerfacecolor='k',linestyle='--',linewidth=3)

#Show legend at the lower right location.
plt.legend(loc='lower right')

#X label name = Month Number
plt.xlabel('Month number')
#Y label name = Sold units number
plt.ylabel('Sold units number')
plt.xticks(months)
plt.yticks([100000, 200000, 300000, 400000, 500000])

#Title
plt.title('Company profit per month')

plt.show()
```

```python
#4.Read all product sales data and show it using a multiline plot Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
#Getting data and converting to List
faceCrem = csv_data ['facecream'].tolist()
faceWash = csv_data ['facewash'].tolist()
toothPaste = csv_data ['toothpaste'].tolist()
bathingsoap = csv_data ['bathingsoap'].tolist()
shampoo = csv_data ['shampoo'].tolist()
moisturizer = csv_data ['moisturizer'].tolist()

#Ploting the graph
plt.plot(months, faceCrem,label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(months, faceWash,label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(months, toothPaste,label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(months, bathingsoap,label = 'Bathingsoap Sales Data', marker='o', linewidth=3)
plt.plot(months, shampoo,label = 'Shampoo Sales Data', marker='o', linewidth=3)
plt.plot(months, moisturizer,label = 'Moisturizer Sales Data', marker='o', linewidth=3)

plt.title('Sales data')
plt.legend(loc='upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

plt.xticks(months)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

plt.show()
print('Analysis : The plot shows Sales data of each product.Lowest sale was Moisturizer and highest was Bathingsoap,The sales of Face wash and Moisturizer are same.' ) 
```

```python
#5.Read toothpaste sales data of each month and show it using a scatter plot Also, add a grid in the plot

plt.scatter(months, toothPaste, label ='Tooth paste Sales data')
plt.title('>> toothpaste sales data of each month <<')
plt.xlabel('Month Number')
plt.ylabel('No of sales')
plt.xticks(months)
#grid in the plot
plt.grid(True, linewidth= 1, linestyle="-")
plt.show()
print('Analysis : The plot shows Toothpaste sales data.Toothpaste sales are more in 10,11,12 months')
```

```python
#6.Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
#Color of Bars
c = ['red', 'yellow', 'black', 'blue', 'orange']

plt.bar(months, bathingsoap,color = c)
plt.xlabel('Month Number')
plt.ylabel('sales data of bathing soap')
plt.title('>> sales data of bathing soap of all months <<')
plt.xticks(months)
plt.title('>> sales data of bathing soap <<')
#Saving this plot to File Folder
plt.savefig('sales data of bathing soap.png')
plt.show()
```

```python
#7.Calculate total sale data for last year for each product and show it using a Pie chart.your pie chart output should like this.

all_salesData   = [sum(faceCrem),sum(faceWash),sum(toothPaste),sum(bathingsoap),sum(shampoo),sum(moisturizer)]

plt.pie(all_salesData, labels=['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc='lower right')
plt.axis("equal")
plt.show()
```

```python
#8.Read Bathing soap ,facewash of all months and display it using the Subplot.your output should be like this as mention below


plt.subplot(2, 1, 1)
plt.plot(months, bathingsoap,label = 'bathingsoap Sales Data', marker='o',color='k', linewidth=3)
plt.title('Sales data of Bathing soap')


plt.subplot(2, 1, 2)
plt.plot(months, faceWash,label = 'Face Wash Sales Data', marker='o',color='r', linewidth=3)
plt.title('Sales data of faceWash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(months)

plt.show()
```

```python
#9.Read all product sales data and show it using the stack plot
#Colors
c = ['m', 'c', 'r', 'k', 'g','y']
labels=['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']

for i in range(len(labels)):
    plt.plot([],[],color=c[i], label=labels[i], linewidth=5)
 
plt.stackplot(months, faceCrem, faceWash, toothPaste, bathingsoap, shampoo, moisturizer, colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
